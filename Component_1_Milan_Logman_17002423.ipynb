# **Missing Values**

```
data.isnull().sum()
```

It can be seen above that the attributes 'URN', 'LA', 'GENDER' and 'Rating' do not contain any missing data. With regards to the other attributes' missing data, the UK government website reveals a number of things. Firstly, there are a number of schools that have opened relatively recently so information such as 'Rating', 'PNUMFSM' will not yet be available. Furthermore, there are cases where the school became an academy or changed type during the 2018 to 2019 academic year. In these cases, the 2019 data is found on the former school's page and, therefore, no data exists for 2020. Other instances exist whereby data related to Progress 8 such as 'P8PUP' and 'P8MEA' are not published if the schools are independent schools and independent special schools, or breakdowns by disadvantaged and other pupils for independent schools, independent special schools and non-maintained special schools. There are also schools on the list that do not have any data available for attributes such as 'ATT8SCR' and 'EBACCAPS' because the school or college did not enter any pupils or students for the relevant qualifications.

# **Handling Missing Data**

Deciding on what to do with the missing data depends on what kind of analysis will be carried out on the dataframe and what kind of answers we are looking for. 

# Option 1
The first option is to remove all rows that contain missing data. This would produce a dataframe that contains no missing data but has significantly less rows (schools).

```
data.isnull().any(axis=1).sum()
```

As shown above, there are 3187 rows (schools) that have one or more values missing. This leaves 3209 schools that do not have any data missing. This is just over 50% of the schools which makes omitting rows with empty values a viable option.

# Option 2
Another option is to omit the columns that contain the most missing values such as 'P8_BANDING' and from there, remove the rows that contain missing data. This will produce a dataframe with more schools than the first option but less attributes to compare. For example:

```
df2=data.drop(columns=['PNUMEAL',
'PNUMENGFL',             
'PNUMUNCFL',               
'PNUMFSMEVER',           
'PNUMFSM',                 
'PERCTOT',            
'PPERSABS10',           
'TPUP',            
'ATT8SCR',             
'P8PUP',                
'P8MEA',                
'P8CILOW',              
'P8CIUPP',              
'PTL2BASICS_94',          
'PTL2BASICS_95',          
'EBACCAPS',            
'PTEBACC_E_PTQ_EE',       
'PTEBACC_94',          
'PTEBACC_95',          
'PT5EM_94',           
'P8_BANDING',
'OVERALL_DESTPER',
'NOT_SUSTAINEDPER',
'UNKNOWNPER',
'OVERALL_DESTPER_DIS',
'NOT_SUSTAINEDPER_DIS',
'UNKNOWNPER_DIS'])
```

```
df2
```

```
df2.isnull().any(axis=1).sum()
```

```
df2.dropna()
```

As shown above, by dropping columns that contain significant amounts of missing data, the number of rows (schools) containing missing data decreased from 3187 to 774. This means that there are 5622 schools to compare instead of 3209 (as in Option 1). However, only 9 columns remain meaning there are less attributes to compare.

# Option 3

Finally, if it is necessary for all schools to be compared, every column containg missing data can be removed such that the dataframe is reduced to the attributes 'URN', 'LA', 'GENDER' and 'Rating'.

```
df3=data.dropna(axis=1)
```

```
df3
```

This means that no rows (schools) need to be removed so all 6396 schools can be compared. However, only the 4 aforementioned attributes can be compared across the schools. 

In this instance, it would be prudent to go with option 1. Maintaining all the attributes will be useful for extensive analysis of each school and allows many different types of comparisons to be made. Although the dataframe will be made smaller, more than 50% of the schools will still remain and virtually every local authority is represented which can help give an idea of what might be expected from the schools that have missing data.

# **Sanity Checks**

```
((data1['PNUMEAL'] + data1['PNUMENGFL'] + data1['PNUMUNCFL']) >99.89).all()
```

PNUMEAL, PNUMENGFL, PNUMUNCFL values should all equal 100 across each row. The code above shows that they all equal at least 99.89. The reason for the use of 99.89 is because of rounding and decimal places that may cause some totals to not equal exactly 100. When 99.9 is used, some totals fall below this threshold.
